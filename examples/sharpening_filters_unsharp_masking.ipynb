```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

first in grayscale

```python
guybrush = cv2.imread('guybrush.jpg', 0)

guybrush = cv2.medianBlur(guybrush, 5)
plt.imshow(guybrush, cmap='gray', vmin=0, vmax=255)
```

```python
guybrush_unsharp = cv2.GaussianBlur(guybrush, (5, 5), 0)

plt.imshow(guybrush_unsharp, cmap='gray', vmin=0, vmax=255)
```

```python
guybrush_sharping = cv2.subtract(guybrush, guybrush_unsharp)

plt.imshow(guybrush_sharping, cmap='gray', vmin=0, vmax=255)
```

```python
guybrush_sharp = cv2.add(guybrush, guybrush_sharping)

plt.imshow(guybrush_sharp, cmap='gray', vmin=0, vmax=255)
```

```python
plt.subplot(121)
plt.imshow(guybrush, cmap='gray', vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(guybrush_sharp, cmap='gray', vmin=0, vmax=255)
```

now in color

```python
guybrush = cv2.imread('guybrush.jpg')
guybrush = cv2.cvtColor(guybrush, cv2.COLOR_BGR2RGB)
guybrush = cv2.medianBlur(guybrush, 5)
plt.imshow(guybrush)
```

```python
guybrush_unsharp = cv2.GaussianBlur(guybrush, (5, 5), 0)
guybrush_sharping = cv2.subtract(guybrush, guybrush_unsharp)

plt.imshow(guybrush_sharping)
```

```python
guybrush_sharp = cv2.add(guybrush, guybrush_sharping)

plt.imshow(guybrush_sharp)
```

```python
plt.subplot(121)
plt.imshow(guybrush)
plt.subplot(122)
plt.imshow(guybrush_sharp)
```

use a designed sharpening kernel directly

```python
guybrush = cv2.imread('guybrush.jpg', 0)
guybrush = cv2.medianBlur(guybrush, 5)
```

```python
kernel = np.array([[ 0,  -1,  0], [ -1, 5,  -1],[ 0,  -1,  0]])
kernel
```

```python
guybrush_sharp = cv2.filter2D(src=guybrush, ddepth=cv2.CV_64F, kernel=kernel)

plt.subplot(121)
plt.imshow(guybrush, cmap='gray', vmin=0, vmax=255)
plt.subplot(122)
plt.imshow(guybrush_sharp, cmap='gray', vmin=0, vmax=255)
```
