```python
import cv2 as cv2
import matplotlib.pyplot as plt
```

```python
r2d2 = cv2.imread('r2d2.jpg')
r2d2_gray= cv2.cvtColor(r2d2, cv2.COLOR_BGR2GRAY)
r2d2 = cv2.cvtColor(r2d2, cv2.COLOR_BGR2RGB)
```

```python
sift = cv2.SIFT_create()
keypoints = sift.detect(r2d2_gray, None)
```

```python
keypoints
```

```python
r2d2_with_keypoints = cv2.drawKeypoints(r2d2, keypoints, 0, (255, 0, 0))
```

```python
plt.imshow(r2d2)
```

```python
plt.imshow(r2d2_with_keypoints)
```

```python
r2d2_with_keypoints = cv2.drawKeypoints(r2d2, keypoints, 0, (255, 0, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(r2d2_with_keypoints)
```
