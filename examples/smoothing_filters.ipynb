```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
guybrush = cv2.imread('guybrush.jpg')
```

```python
guybrush.shape
```

```python
guybrush[100:150, 100:150]
```

```python
guybrush = cv2.cvtColor(guybrush, cv2.COLOR_BGR2RGB)
plt.imshow(guybrush)
```

normalized box filter

```python
guybrush_blur = cv2.blur(guybrush, (5, 5))

plt.subplot(121)
plt.imshow(guybrush)
plt.subplot(122)
plt.imshow(guybrush_blur)
```

same by manually setting the kernel

```python
kernel = np.ones((5, 5), np.float32) / 25
guybrush_box = cv2.filter2D(src=guybrush, ddepth=-1, kernel=kernel)
# setting ddepth to -1 ensures the output image has the same depth as the input image

plt.subplot(131)
plt.imshow(guybrush)
plt.subplot(132)
plt.imshow(guybrush_blur)
plt.subplot(133)
plt.imshow(guybrush_box)
```

```python
guybrush_gaussian = cv2.GaussianBlur(guybrush, (5, 5), 0)

plt.subplot(131)
plt.imshow(guybrush)
plt.subplot(132)
plt.imshow(guybrush_blur)
plt.subplot(133)
plt.imshow(guybrush_gaussian)
```

```python
guybrush_median = cv2.medianBlur(guybrush, 5)

plt.subplot(121)
plt.imshow(guybrush)
plt.subplot(122)
plt.imshow(guybrush_median)
```

```python
guybrush_bilateral = cv2.bilateralFilter(guybrush, 9, 75, 75)

plt.subplot(131)
plt.imshow(guybrush)
plt.subplot(132)
plt.imshow(guybrush_gaussian)
plt.subplot(133)
plt.imshow(guybrush_bilateral)
```
