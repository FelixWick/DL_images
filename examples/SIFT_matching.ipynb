```python
import cv2 as cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('almanac.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

img2 = cv2.imread('desk.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
```

initiate SIFT

```python
sift = cv2.SIFT_create()
```

```python
keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)
```

brute-force matcher

```python
bf = cv2.BFMatcher(cv2.NORM_L2)
matches = bf.match(descriptors1, descriptors2)
```

```python
len(keypoints1)
```

```python
len(keypoints2)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
```

```python
matches[:10]
```

```python
image_matched = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS, matchesThickness=3)
```

```python
plt.imshow(image_matched)
```
