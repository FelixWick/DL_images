```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
guybrush = cv2.imread('guybrush.jpg')
guybrush = cv2.cvtColor(guybrush, cv2.COLOR_BGR2RGB)
```

```python
guybrush_gray = cv2.cvtColor(guybrush, cv2.COLOR_BGR2GRAY).astype(np.float32)
```

arguments:
- blockSize - size of neighbourhood considered for corner detection
- ksize - aperture parameter of Sobel derivative used
- k - free parameter in Harris detector response function, see slides

```python
harris_response = cv2.cornerHarris(guybrush_gray, 2, 3, 0.04)
harris_response[100:200, 100:200]
```

```python
plt.imshow(harris_response, cmap='gray', vmin=0, vmax=255)
```

```python
# used threshold
guybrush[harris_response > 0.01 * harris_response.max()] = [0 ,255 , 0]
```
